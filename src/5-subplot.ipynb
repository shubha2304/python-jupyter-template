### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.arange(1, 6)
y = np.array([2, 4, 6, 8, 10])
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Line chart
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='blue', label="Line")
axes[0, 0].set_title("Line Chart")
axes[0, 0].legend()

# Second subplot: Bar chart
axes[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axes[0, 1].set_title("Bar Chart")

# Third subplot: Scatter plot
axes[1, 0].scatter(x, y, color='purple')
axes[1, 0].set_title("Scatter Plot")
for i, txt in enumerate(y):
    axes[1, 0].annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Fourth subplot: Pie chart
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'], startangle=140)
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
